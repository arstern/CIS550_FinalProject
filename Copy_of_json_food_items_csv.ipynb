```python
import pandas as pd
def rest_json_to_df(rest_json):
  dishes = []
  for rest in rest_json:
    menu_info = rest['menus']
    rest_id = rest['restaurant_id']
    for menu in menu_info:
      for section in menu['menu_sections']:
        menu_items = section['menu_items']
        for food_item_dict in menu_items:
          food_item_dict['rest_id'] = rest_id
          if any(name.isdigit() for name in food_item_dict['name']):
            num_removed = food_item_dict['name'].split(' ')[1:]
            name = ' '.join(num_removed)
            food_item_dict['name'] = name
          dishes.append(food_item_dict)
  
  df = pd.DataFrame.from_dict(dishes)
  df = df.drop(['pricing'], axis=1)
  return df
```

```python
df = rest_json_to_df(data)
```
