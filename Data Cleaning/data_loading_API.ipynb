```python
n_nyc = 0
for page_num in range(1,1828):
  g2 = 'https://api.documenu.com/v2/restaurants/search/fields?state=NY&exact=True&size=30&page=' + str(page_num)
  r2 = requests.get(g2, headers=headers).json()
  for location in r2['data']:
    if location['address']['city'].lower() == 'new york':
      n_nyc += 1
  if page_num % 100 == 0:
    print(page_num, n_nyc)
```
